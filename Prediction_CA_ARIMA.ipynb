```python
train_data = df['Quarter Turnover Market Base Business'][:32]
test_data = df['Quarter Turnover Market Base Business'][32:]

# define model
my_order = (0,0,0) #(000,1014)(2, 0, 2, 4)
my_seasonal_order = (2, 0, 2, 4)
model = SARIMAX(train_data, order=my_order, seasonal_order=my_seasonal_order)
model_fit = model.fit()
print(model_fit.summary())

predictions = model_fit.forecast(len(test_data))
residuals=predictions-test_data

#avoir
plt.figure(figsize=(20,5))
plt.title('Courbe du chiffre d\'affaires du marché  trimestriel et des prédictions')
plt.plot(df['Trimestre de Date'],df['Quarter Turnover Market Base Business'])
plt.plot(df['Trimestre de Date'][-13:],predictions)
plt.xticks(rotation=90)
plt.ylabel('Chiffre d\'affaires trimestriel en €')
plt.xlabel('Trimestre')
plt.legend(('Data', 'Prédictions'), fontsize=10,loc="lower left")

# plt.figure(figsize=(10,4))
# plt.plot(df['Quarter Turnover Market Base Business'])
# plt.plot(predictions)
# plt.legend(('Data', 'Predictions'), fontsize=10,loc="lower left")
```

```python
print('Mean Absolute Percent Error:', round(np.mean(abs(residuals/test_data)),4))
print('Root Mean Squared Error:', np.sqrt(np.mean(residuals**2)))
```

### Maintenant le test est bon on entraine le modele sur tout le signal et on fait la prédiction Quarter sur 5 ans

```python
train_data = df['Quarter Turnover Market Base Business'][:44]
test_data = df['Quarter Turnover Market Base Business'][40:]
```

```python
my_order = (0,0,0) #(000,1014)(2, 0, 2, 4)
my_seasonal_order = (2, 0, 2, 4)
# define model
model = SARIMAX(train_data, order=my_order, seasonal_order=my_seasonal_order)
model_fit = model.fit()
```

```python
print(model_fit.summary())
```

```python
# predictions = model_fit.forecast(len(test_data))
# predictions = model_fit.forecast(21)
# print(predictions)

predictions = model_fit.get_forecast(steps=21)
print(predictions.predicted_mean)        # forecast
(predictions.conf_int(alpha=0.0125))
# print(predictions)
# predictions = pd.Series(predictions, index=test_data.index+4)
# print(predictions)
# residuals = test_data - predictions
```

```python
plt.figure(figsize=(10,4))
plt.title('Quarter Turnover Market Prediction')
plt.plot(df['Quarter Turnover Market Base Business'])
plt.plot(predictions.predicted_mean)
# plt.plot(predictions.conf_int(alpha=0.05))
plt.fill_between(predictions.predicted_mean.index, predictions.conf_int(alpha=0.05)['lower Quarter Turnover Market Base Business'], predictions.conf_int(alpha=0.05)['upper Quarter Turnover Market Base Business'], color='grey',alpha=.2)
plt.legend(('Data', 'Predictions','95% Confidence'), fontsize=10,loc="lower left")
```

### Prediction chiffre affaire sur 5ans

```python
df_new=df[['Trimestre de Date','MAT Turnover Market Base Business','Quarter Turnover Market Base Business']]
df_new['upper_quarter']=df_new['Quarter Turnover Market Base Business']
df_new['lower_quarter']=df_new['Quarter Turnover Market Base Business']

df_new['MAT Turnover Market Base Business'][13:16]=df['verif_mk'][13:16]

df_new.head()
```

```python
dictio={45:'2021 T1',46:'2021 T2',47:'2021 T3',48:'2021 T4',49:'2022 T1',50:'2022 T2',51:'2022 T3',52:'2022 T4',
            53:'2023 T1',54:'2023 T2',55:'2023 T3',56:'2023 T4',57:'2024 T1',58:'2024 T2',59:'2024 T3',60:'2024 T4',
            61:'2025 T1',62:'2025 T2',63:'2025 T3',64:'2025 T4'}

for key,value in dictio.items():
    new_row = {'Trimestre de Date':value, 'MAT Turnover Market Base Business':-1, 'Quarter Turnover Market Base Business':predictions.predicted_mean[key],'upper_quarter':predictions.conf_int(alpha=0.05)['upper Quarter Turnover Market Base Business'][key],'lower_quarter':predictions.conf_int(alpha=0.05)['lower Quarter Turnover Market Base Business'][key]}
    df_new = df_new.append(new_row, ignore_index=True)

df_new['upper_mat']=df_new['MAT Turnover Market Base Business']
df_new['lower_mat']=df_new['MAT Turnover Market Base Business']
df_new.tail()
```
